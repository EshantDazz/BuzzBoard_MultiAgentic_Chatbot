```python
import pandas as pd

# Load the Excel file
file_path = "Data/Sample Data for the Model.xlsx"  # Replace with the actual path to your file
data_sheet_full = pd.read_excel(file_path, header=None)

# Identify all columns related to "All Signals"
# Assumes the second row (index 1) contains column headers
all_signals_columns = data_sheet_full.columns[
    data_sheet_full.iloc[1].str.contains("All Signals", case=False, na=False)
]

# Extract rows for these columns
all_signals_data = data_sheet_full.loc[:, all_signals_columns]

# Save the extracted data to a new file for review
output_file_path = "all_signals_data.xlsx"  # Save it as Excel
all_signals_data.to_excel(output_file_path, index=False)

print(f"Data extracted and saved to: {output_file_path}")
```
