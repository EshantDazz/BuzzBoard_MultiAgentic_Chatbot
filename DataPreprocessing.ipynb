## **Filtering Data based on the Json Data provided**

```python
import pandas as pd

# Load the Excel file
file_path = "Data/Sample Data for the Model.xlsx"  # Replace with the actual path to your file
data_sheet_full = pd.read_excel(file_path, header=None)

# Identify all columns related to "All Signals"
# Assumes the second row (index 1) contains column headers
all_signals_columns = data_sheet_full.columns[
    data_sheet_full.iloc[1].str.contains("All Signals", case=False, na=False)
]

# Extract rows for these columns
all_signals_data = data_sheet_full.loc[:, all_signals_columns]

# Save the extracted data to a new file for review
output_file_path = "all_signals_data.xlsx"  # Save it as Excel
all_signals_data.to_excel(output_file_path, index=False)

print(f"Data extracted and saved to: {output_file_path}")
```

### **Storing the DataFrame in the form of Json so that I can send that as part of the prompt**

```python



big_json=df.to_json(orient="records", indent=4)
# Save the JSON to a file
with open("big_output.json", "w") as json_file:
    json_file.write(big_json)

print("JSON file saved as 'big_output.json'")
```

```python
import json

# Path to the input JSON file
input_json_file = 'big_output.json'

# Path to the output Python file
output_py_file = 'big_data.py'

# Load the JSON data
with open(input_json_file, 'r') as json_file:
    data = json.load(json_file)

# Convert JSON nulls to Python None explicitly during writing
with open(output_py_file, 'w') as py_file:
    py_file.write("# Automatically generated Python file\n\n")
    py_file.write("big_data = ")
    # Use json.dumps with default and ensure_ascii=False for proper conversion
    py_file.write(json.dumps(data, indent=4).replace("null", "None"))
    py_file.write("\n")
```
